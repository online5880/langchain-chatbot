```python
df = pd.read_csv('extracted_data.csv')
```

```python
def merge_all_answers(rows):
    answers = []
    for _, row in rows.iterrows():
        # For each row, append answers from all columns (answer_1 to answer_6)
        for answer in ['answer_1', 'answer_2', 'answer_3', 'answer_4', 'answer_5', 'answer_6']:
            if pd.notnull(row[answer]):  # Only add non-null answers
                answers.append(row[answer])
    # Return all answers as a single comma-separated string
    return ', '.join(answers)

# Group by 'question' and apply the merging function
grouped_answers = df.groupby('question').apply(merge_all_answers).reset_index(name='answers')

# Save the grouped results to a new CSV
output_file_grouped = '/mnt/data/final_grouped_answers_by_question.csv'
```
